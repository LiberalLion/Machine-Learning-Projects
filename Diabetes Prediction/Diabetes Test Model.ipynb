```python
import joblib
import pandas as pd
model = joblib.load('diabetes_rf_model.pkl')

df = pd.read_csv
new_data = pd.read_csv('diabetes_test.csv')
```

```python
new_data.head()
```

```python
predict = model.predict(new_data)
predict
```

```python
new_data['Predicted Outcome'] = predict
new_data.head()
```

```python
df2 = pd.DataFrame(new_data)
df2.head()
```

```python
df2.to_csv('predicted_outcomes.csv')
```

```python
# Changing 1 to positive and 0 to negative
new_data.loc[new_data['Predicted Outcome']==1,'Predicted Outcome'] = 'Positve'

# Changing 1 to positive and 0 to negative
new_data.loc[new_data['Predicted Outcome']==0,'Predicted Outcome'] = 'Negative'
```

```python
new_data.head()
```

```python
df3 = pd.DataFrame(new_data)
```

```python
df3.to_csv('Predictions.csv')
```
