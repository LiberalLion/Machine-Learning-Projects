# Movie Recommendation System
In this project I will use pearson correlation method to create a movie recommendation system with collaborative filtering.

Import the required libraries

```python
import pandas as pd
```

```python
ratings = pd.read_csv('userRatings.csv')
ratings.head()
```

We can see that there are 2269 columns meaning there are 2269 movies in the dataset.

```python
ratings.sample(5)
```

```python
ratings.info()
```

There are 610 ratings for 2269 movies in this dataset.

```python
ratings.describe().T
```

We can see that the ratings range from 0.0 to 5.0

```python
def standardize(row):
    new_row = (row-row.mean())/(row.max()-row.min())
    return new_row

ratings_std = ratings.apply(standardize)
ratings_std
```

```python
corrMatrix = ratings.corr(method='pearson')
```

The method above calls the function .corr and we set the method to pearson for the pearson correlation method.

```python
corrMatrix
```

We can now see the corresponding correlation values in the data. A higher rating between 2 movies mean they are highly correlated and are similar therefore we can recommend that movie to the user.

```python
corrMatrix_df = pd.DataFrame(corrMatrix, index = ratings.columns, columns = ratings.columns)
corrMatrix_df
```

```python
def get_similarity(movie_name, rating):
    similar_score = corrMatrix_df[movie_name]*(rating-2.5)
    
    similar_score = similar_score.sort_values(ascending = False)
    
    return similar_score
```

```python
get_similarity("12 Years a Slave (2013)",5)
```

```python
action_lover = [("Zootopia (2016)", 1), ("Zoolander (2001)",5), ("xXx (2002)",5)]
```

```python
similar_scores = pd.DataFrame()

for movie, rating in action_lover:
    similar_scores = similar_scores.append(get_similarity(movie,rating), ignore_index = True)
    
similar_scores.head()
```

```python
similar_scores.sum().sort_values(ascending = False)
```

We can see that the user gave ratings of 5 to xXx and Zoolander which are action and comedy movies and gave a low rating to Zootopia which is an animated movie. Therefore movies like Moana, Coco etc will not be recommended to the user. Movies like Die another day, school of rock, S.W.A.T can be recommended to the user because their similarity ratings are high and they're related(action and comedy movies).
