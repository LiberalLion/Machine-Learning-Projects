```python
import pandas as pd
dataset = [['Milk', 'Egg', 'Bread', 'Butter'],
           ['Milk', 'Butter', 'Egg', 'Ketchup'],
           ['Bread', 'Butter', 'Ketchup'],
           ['Milk', 'Bread', 'Butter'],
           ['Cookies', 'Bread', 'Butter'],
           ['Milk', 'Bread', 'Butter', 'Cookies'],
           ['Milk', 'Cookies'],
           ['Milk', 'Bread', 'Butter'],
           ['Egg', 'Bread', 'Butter', 'Cookies'],
           ['Milk', 'Bread', 'Butter'],
           ['Milk', 'Bread', 'Cookies', 'Ketchup']]
```

```python
dataset
```

```python
df1 = pd.DataFrame(dataset)
df1
```

```python
 from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()

te_ary = te.fit(dataset).transform(dataset)
```

```python
te_ary
```

```python
df = pd.DataFrame(te_ary, columns = te.columns_)
df.head()
```

```python
# We can see that from the first row we have Bread, Buttern Egg and Milk set to True and the rest set to false. Comparing with the initial dataset, The first row doesn't have cookies and ketchup so that's why they are set to false.
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(df, min_support = 0.3, use_colnames = True)

frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

result = association_rules(frequent_itemsets, metric = 'confidence', min_threshold = 0.6)

result
```

```python
result1 = result[['antecedents', 'consequents', 'support', 'confidence', 'lift']]

result1
```

```python
result1.sort_values('lift', ascending = False)
```
